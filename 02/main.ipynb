```python
reports = []

with open('input.txt') as f:
    for line in f:
        reports.append([int(n) for n in line.strip().split(" ")])

def is_safe(report):
    print(report, end=" ")
    inc_dec_type = None
    for i, n in enumerate(report[1:]):
        if abs(n - report[i]) > 3 or n == report[i]:
            print("unsafe")
            return 0
        elif n - report[i] > 0:
            if inc_dec_type == "decreasing":
                print("unsafe")
                return 0
            inc_dec_type = "increasing"
        elif n - report[i] < 0:
            if inc_dec_type == "increasing":
                print("unsafe")
                return 0
            inc_dec_type = "decreasing"
            
    print("safe")
    return 1

safe_reports = 0
for report in reports:
    safe_reports += is_safe(report)

print(safe_reports)
```

```python
reports = []

with open('input.txt') as f:
    for line in f:
        reports.append([int(n) for n in line.strip().split(" ")])

safe_reports_count = 0
for report in reports:
    if is_safe(report):
        safe_reports_count += 1
        continue
    for del_i in range(0, len(report)):
        subreport = report[:del_i] + report[del_i+1:]
        if (is_safe(subreport)):
            safe_reports_count += 1
            break

print(safe_reports_count)
```
