```python
with open('input.txt') as f:
    lines = f.readlines()

matrix = []
for line in lines:
    matrix.append(list(line.strip()))

# short view
print(f'    {'    '.join([str(x) for x in range(10)])}')
for r, line in enumerate(matrix[:10]):
    print(f'{r} {line[:10]}')
```

```python
# since X in XMAS is unique, we can find all coordinates of 'X'

x_coords = []
for i, line in enumerate(matrix):
    for j, char in enumerate(line):
        if char == 'X':
            x_coords.append((i, j))

print(len(x_coords), x_coords)
```

```python
# count in all directions from X
# where 'M', 'A' and 'S' are in sequence in that direction, increment the counter

counter = 0

def is_xmas_south(i, j):
    if i + 3 >= len(matrix):
        return False
    return matrix[i+1][j] == 'M' and matrix[i+2][j] == 'A' and matrix[i+3][j] == 'S'

def is_xmas_north(i, j):
    if i - 3 < 0:
        return False
    return matrix[i-1][j] == 'M' and matrix[i-2][j] == 'A' and matrix[i-3][j] == 'S'

def is_xmas_east(i, j):
    if j + 3 >= len(matrix[0]):
        return False
    return matrix[i][j+1] == 'M' and matrix[i][j+2] == 'A' and matrix[i][j+3] == 'S'

def is_xmas_west(i,j):
    if j - 3 < 0:
        return False
    return matrix[i][j-1] == 'M' and matrix[i][j-2] == 'A' and matrix[i][j-3] == 'S'

def is_xmas_north_east(i, j):
    if i - 3 < 0 or j + 3 >= len(matrix[0]):
        return False
    return matrix[i-1][j+1] == 'M' and matrix[i-2][j+2] == 'A' and matrix[i-3][j+3] == 'S'

def is_xmas_north_west(i, j):
    if i - 3 < 0 or j - 3 < 0:
        return False
    return matrix[i-1][j-1] == 'M' and matrix[i-2][j-2] == 'A' and matrix[i-3][j-3] == 'S'

def is_xmas_south_east(i, j):
    if i + 3 >= len(matrix) or j + 3 >= len(matrix[0]):
        return False
    return matrix[i+1][j+1] == 'M' and matrix[i+2][j+2] == 'A' and matrix[i+3][j+3] == 'S'

def is_xmas_south_west(i, j):
    if i + 3 >= len(matrix) or j - 3 < 0:
        return False
    return matrix[i+1][j-1] == 'M' and matrix[i+2][j-2] == 'A' and matrix[i+3][j-3] == 'S'




for i, j in x_coords:
    if is_xmas_south(i, j):
        counter += 1
    if is_xmas_north(i, j):
        counter += 1
    if is_xmas_east(i, j):
        counter += 1
    if is_xmas_west(i, j):
        counter += 1
    if is_xmas_north_east(i, j):
        counter += 1
    if is_xmas_north_west(i, j):
        counter += 1
    if is_xmas_south_east(i, j):
        counter += 1
    if is_xmas_south_west(i, j):
        counter += 1
    
print(f'Answer: {counter}')
```
