```python
connections = {}
with open('input.txt') as f:
    for line in f:
        a, b = line.strip().split('-')
        if a not in connections:
            connections[a] = []
        connections[a].append(b)
        if b not in connections:
            connections[b] = []
        connections[b].append(a)

connections
```

Let  k  represent the number of passes over  C . For the  k-th pass:


$I_j^{(k)} \gets I_j^{(k-1)} \cup \{C_i\}, \quad \text{if } I_j^{(k-1)} \subseteq C_i^N \text{ for some } I_j$


If no such  $I_j^{(k)}$  exists, create a new clique:

$I_{j+1}^{(k)} = \{C_i\}$


At the end of the process,  $\mathcal{I} = \{I_1, I_2, \dots, I_m\}$  contains all maximal cliques found.

Start with  $\mathcal{I} = \{ \}$ (an empty set of internets).


For each  $C_i \in C$ :
a. Check if there exists  $I_j \in \mathcal{I}$  such that  $I_j \subseteq C_i^N$ .
b. If such  $I_j$  exists, add  $C_i$  to $I_j$ :

$I_j \gets I_j \cup \{C_i\}$


```python
def increase_clique(I: list[list[str]], do_print=False):
    # base case
    if len(I) == 0:
        I += [[x] for x in connections.keys()]

    has_changed = False
    for c in connections.keys():
        for i in I:
            if do_print:
                print(f"checking if I: {i} is in {c}'s connections")
                print(connections[c])
            is_in = all([x in connections[c] for x in i])
            if do_print:
                print(is_in)

            if is_in:
                if do_print:
                    print(f"## all elements of {i} are in {c}'s connections ##")
                    print(f"## adding {c} to {i} ##")
                i.append(c)
                has_changed = True

    return has_changed

I: list[list[str]] = []
has_changed = True
while has_changed:
    has_changed = increase_clique(I, do_print=False)

max_clique = max(I, key=len)
max_clique.sort()
",".join(max_clique)
```
