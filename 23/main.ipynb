# LAN party

- pair-wise connection (`kh` to `tc`) between computers
- connections are bidirectional
- letterings of each computer is not arbitrary (can be used for search)
- find all sets of three computers where each computer is connected to the other two
- filter by the ones that have at least one computer with the letter `t` in it

```python
def find_three_computers(connections):
    three_computers = []
    for computer, children in connections.items():
        for c in children:
            for gc in connections[c]:
                for ggc in connections[gc]:
                    if computer == gc or c == ggc:
                        continue
                    if computer == ggc:
                        three_computers.append([computer, c, gc])

            
    three_computers = [sorted(x) for x in three_computers]
    three_computers = set([tuple(x) for x in three_computers])

    return three_computers

connections = {}
with open('example.txt') as f:
    for line in f:
        a, b = line.strip().split('-')
        if a not in connections:
            connections[a] = []
        connections[a].append(b)
        if b not in connections:
            connections[b] = []
        connections[b].append(a)

three_computers = find_three_computers(connections)
assert len(three_computers) == 12

print(three_computers)

def find_starts_with_t():
    t_count = 0
    for three_com in three_computers:
        print(three_com, end="")

        if any([x[0] == "t" for x in three_com]):
            print(" has some that start with t")
            t_count += 1
        else:
            print()

    return t_count

t_count = find_starts_with_t()
print(t_count)
assert t_count == 7, t_count
```
