```python
tokens = []
design_wishes = []
n = 0
with open('input.txt') as f:
    for line in f:
        if n == 0:
            tokens.extend([t.strip() for t in line.split(",")])
        else:
            if line.strip() != "":
                design_wishes.append(line.strip())

        n += 1

token_characterset = set("".join(tokens))
print(tokens)
print(design_wishes)
```

```python
memory = {}

def rec_search(design: str, tokens: list[str]) -> bool:
    if design == "":
        return True
    
    # check if the result is already in the memoization table
    if design in memory:
        return memory[design]
    
    ndesigns = []
    for i in range(len(tokens)):
        if not design.startswith(tokens[i]):
            # if the design does not start with the token, then it is not possible to make the design
            continue

        t_count = design.count(tokens[i])

        if t_count > 0:
            ndesigns.append(rec_search(design.replace(tokens[i], "", 1), tokens))
        else:
            continue
    
    # if any of the ndesigns is True, then the design can be made
    is_possible = any(ndesigns)
    memory[design] = is_possible
    return is_possible

count = 0
for design in design_wishes:
    # only relevant tokens need be considered
    rel_tokens = [t for t in tokens if t in design]

    res = rec_search(design, rel_tokens)
    memory[design] = res
    print(f"{design}: {res}")
    if res:
        count += 1

print(f"Number of designs that can be made: {count}")
print(memory)
```
