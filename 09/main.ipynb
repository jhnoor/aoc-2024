```python
disk_map = []
with open('input.txt') as f:
    disk_map = [int(n) for n in f.readline().strip()]

blocks = []
for i in range(len(disk_map)):
    if i % 2 == 0:
        id = i//2
    else:
        id = '.'

    n_blocks = disk_map[i]
    for j in range(n_blocks):
        blocks.append(id)

def pretty_print(blocks):
    print("".join([str(b) for b in blocks]))

def get_size(blocks, i):
    # get size and locations of blocks
    size = 1
    start = i
    end = i
    for j in range(i+1, len(blocks)):
        if blocks[j] == blocks[i]:
            size += 1
            end = j
        else:
            break
    
    return (size, start, end)
        

def size_up(blocks):
    sizes = []
    i = 0
    while True:
        if i >= len(blocks):
            break
        if blocks[i] == '.':
            i += 1
            continue
        size, start, end = get_size(blocks, i)
        sizes.append(
            {
                'id': blocks[i],
                'size': size,
                'start': start,
                'end': end
            }
        )
        i = end+1
    
    return sizes

def get_free(blocks):
    free = []
    i = 0
    while True:
        if i >= len(blocks):
            break
        if blocks[i] == '.':
            size, start, end = get_size(blocks, i)
            free.append(
                {
                    'size': size,
                    'start': start,
                    'end': end
                }
            )
            i = end+1
        else:
            i += 1
    
    return free

def move(block, free_space):
    b_size = block['size']
    b_start = block['start']
    b_end = block['end']
    f_start = free_space['start']

    blocks[f_start:f_start+b_size] = blocks[b_start:b_end+1]
    blocks[b_start:b_end+1] = ['.']*(b_end-b_start+1)

def move_blocks_to_free_space(blocks):
    sizes = size_up(blocks)
    free = get_free(blocks)

    for s in sizes[::-1]:
        possible_moves = [f for f in free if f['size'] >= s['size']]
        if any(possible_moves):
            # if move is left to right, cancel
            if s['start'] < possible_moves[0]['start']:
                continue
            move(s, possible_moves[0])
            free = get_free(blocks)


move_blocks_to_free_space(blocks)
b_sum = sum([i*j for i, j in zip(blocks, range(len(blocks))) if i != '.'])
print(b_sum)
```
